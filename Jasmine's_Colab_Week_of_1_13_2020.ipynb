```
import numpy as np
import mpl_toolkits.mplot3d.axes3d as p3
import matplotlib.pyplot as plt
import colorsys
from PIL import Image
import pandas as pd
from urllib.request import urlopen

# (1) Import the file to be analyzed!
img_file = Image.open(urlopen("https://hips.hearstapps.com/hmg-prod.s3.amazonaws.com/images/blueivy-beyonce-1563567558.jpg"))
img = img_file.load()   # this is called a PIXEL ACCESS OBJECT

# (2) Construct a blank matrix representing the pixels in the image
[xs, ys] = img_file.size
max_intensity = 100
hues = {}

# (3) Examine each pixel in the image file and add it to an array
r_ = []
g_ = []
b_ = []
colours = []

for x in range(0, xs):
  for y in range(0, ys):
    # ( )  Get the RGB color of the pixel
    [r, g, b] = img[x, y]

    # ( )  Normalize pixel color values
    r /= 255.0
    g /= 255.0
    b /= 255.0

    r_.append(r)
    g_.append(g)
    b_.append(b)
    colours.append([r, g, b])
    
    img[x,y] = (int(r*255), 0, 0)   # note by messing with the pixel access object, I'm changing the image!
    
# (4)   Plot the array!
fig = plt.figure()
ax = p3.Axes3D(fig)
ax.scatter(r_, g_, b_, s=5, c=colours, lw=0)

ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')
# fig.add_axes(ax)
plt.show()
```
